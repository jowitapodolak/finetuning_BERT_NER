# Langauge Similarity : Comparison of mBERT and XLMR

```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import FormatStrFormatter
import seaborn as sns
from statistics import mean, variance
%matplotlib inline
```

1) Monolingual results:

```python
monolingual_mBERT = pd.read_csv('mBERT/Monolingual_results_mBERT.csv', sep=',')
monolingual_mBERT['model']='mBERT'
monolingual_XLMR = pd.read_csv('XLMR/Monolingual_results_XLMR.csv', sep=',')
monolingual_XLMR['model']='XLMR'
monolingual_mBERT.head()
```

```python
temp1 = monolingual_mBERT[['LAN_1','mean','standard dev','model']]
temp2 = monolingual_XLMR[['LAN_1','mean','standard dev','model']]
monolingual = pd.concat([temp1, temp2]).sort_values(by=['mean'], ascending=False)
monolingual.head()
```

```python
N = len(monolingual_mBERT['LAN_1'].unique())
ind = np.arange(N)


fig, ax = plt.subplots(figsize=(30,5))
#plt.yscale('log')
ax.set_ylim(ymax = 100, ymin = 70)

ax = sns.set(style="whitegrid")
ax = sns.despine(left=True)
ax = sns.barplot(x="LAN_1", y="mean", hue="model",data=monolingual)
ax.set_xlabel('Language')
#ax.set_title('Monolingual Results mBERT', fontsize=14, fontweight='bold')
ax.set_ylabel('Average F1 Score in %')

for p in ax.patches:
        ax.annotate('{}'.format(round(p.get_height(),1)), (p.get_x()+0.3, p.get_height()+0.6),
                    ha='center', va='bottom',
                    color= 'black',fontsize=12)

plt.savefig('Results_Monolingual_bothModels.png', bbox_inches='tight')
plt.show()
```

```python
lanDist_mBERT = pd.read_csv('mBERT/Shared_Features_mBERT.csv', sep=',')
lanDist_mBERT['model']='mBERT'
lanDist_XLMR = pd.read_csv('XLMR/Shared_Features_XLMR.csv', sep=',')
lanDist_XLMR['model']='XLMR'
lanDist_mBERT.head()
```

```python
temp1 = lanDist_mBERT[['common_features','LAN_1','LAN_2','mean','standard dev','model']]
temp2 = lanDist_XLMR[['common_features','LAN_1','LAN_2','mean','standard dev','model']]
lanDist = pd.concat([temp1, temp2])
lanDist.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax = sns.set(style="whitegrid")
ax = sns.despine(left=True)
#err_style="bars"
ax = sns.lineplot(x="common_features", y="mean",hue="model", style="model",markers=["D","X"], dashes=False, data=lanDist)
#ax.set_title('Cross-lingual F1-Scores and shared typological ES features in mBERT', fontsize=14, fontweight='bold')
ax.set_ylabel('F1-Score in %')
#plt.yscale('log')
#plt.ylim(ymax = 100, ymin = 70)
ax.set_xlabel('Shared ES features between language pair')

plt.savefig('Results_LanDist_models.png', bbox_inches='tight')

plt.show()
```
