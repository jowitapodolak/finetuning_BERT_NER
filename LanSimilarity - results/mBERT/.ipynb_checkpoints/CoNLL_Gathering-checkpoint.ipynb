# Langauge Similariy : Analysing Results of mBERT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import FormatStrFormatter
import seaborn as sns
from statistics import mean, variance
%matplotlib inline
```

1) Read in results:

```python
lan_results_raw = pd.read_csv('LanSimilarity_results_mBERT.csv', sep=',')
lan_results_raw.head()
```

2) Data Cleaning: Filter for f1 scores + mean/variance calculation of f1 score

```python
three_it = lan_results_raw[lan_results_raw['f1_4'] == 0]
three_it.head()
```

```python
lan_results_f1 = lan_results_raw[['DIST','LAN_1', 'LAN_2', 'f1_1','f1_2','f1_3','f1_4']]
    
lan_results_f1 = lan_results_f1.reindex(columns = ['DIST','LAN_1', 'LAN_2','f1_1','f1_2','f1_3','f1_4','mean','variance', 'standard dev'])

lan_results_f1['f1_1'] = lan_results_f1['f1_1'].apply(lambda x: x*100)
lan_results_f1['f1_2'] = lan_results_f1['f1_2'].apply(lambda x: x*100)
lan_results_f1['f1_3'] = lan_results_f1['f1_3'].apply(lambda x: x*100)
lan_results_f1['f1_4'] = lan_results_f1['f1_4'].apply(lambda x: x*100)

for index, row in lan_results_f1.iterrows():
    
    if row['f1_4'] == 0:
        lan_results_f1.at[index,'mean'] = np.mean([row['f1_1'],row['f1_2'],row['f1_3']])
        lan_results_f1.at[index,'variance'] = np.var([row['f1_1'],row['f1_2'],row['f1_3']])
        lan_results_f1.at[index,'standard dev'] = np.std([row['f1_1'],row['f1_2'],row['f1_3']])
    else:
        lan_results_f1.at[index,'mean'] = np.mean([row['f1_1'],row['f1_2'],row['f1_3'],row['f1_4']])
        lan_results_f1.at[index,'variance'] = np.var([row['f1_1'],row['f1_2'],row['f1_3'],row['f1_4']])
        lan_results_f1.at[index,'standard dev'] = np.std([row['f1_1'],row['f1_2'],row['f1_3'],row['f1_4']])

lan_results_f1.head()
```

3) Analysing monolingual results - mean/ variance

```python
monolingual = lan_results_f1[(lan_results_f1['DIST'] == 0.0) & (lan_results_f1['LAN_1'] == lan_results_f1['LAN_2'])]
monolingual = monolingual.sort_values(by = "mean", ascending = False)
len(monolingual)
```

```python
# language length
language_length = pd.read_csv('../Languages_length.csv', sep=';')
language_length['no of lines'] = language_length['no of lines'].apply(lambda x : int(float(x.replace(',','.')))/100000)
language_length = language_length.rename(columns={'Language abbreviation': 'LAN_1'})
language_length
language_length_mod = pd.merge(language_length, monolingual, on='LAN_1').sort_values(by='mean', ascending=False)
language_length_mod = language_length_mod[language_length_mod["mean"] != ""]
len(language_length_mod)
#language_length_mod
```

```python
sns.lineplot(x='LAN_1', y='no of lines', data=language_length_mod, sort=False)
```

```python
#plot the results

# How to integrate lineplot in barplot:
# https://towardsdatascience.com/combo-charts-with-seaborn-and-python-2bc911a08950

x = monolingual['LAN_1'].values
N = len(monolingual['LAN_1'].unique())
ind = np.arange(N)
width = 0.35 
y_mean = monolingual['mean'].values
y_std = monolingual['standard dev'].values
 

fig, ax1 = plt.subplots(figsize=(20,3))
#ax1.set_title('Monolingual Results mBERT', fontsize=14, fontweight='bold')
ax1.set_ylabel('F1-Score: Average in %')
#plt.yscale('log')
ax1.set_xlabel('Language')
ax1 = sns.set(style="whitegrid")
ax1 = sns.despine(left=True)
ax1 = sns.barplot(x, y_mean, color='c', saturation=0.7, yerr=y_std)
ax2 = ax1.twinx()
color = 'tab:green'
ax2 = sns.lineplot(x='LAN_1', y='no of lines', data=language_length_mod, sort=False, color=color, linewidth=2)
ax2.set_ylabel('Datasize: No. of lines in T')
ax2.set_ylim(ymax = 500)
ax2.tick_params(axis='y', color=color)

for p in ax1.patches:
        ax1.annotate('{}'.format(round(p.get_height(),1)), (p.get_x()+0.3, p.get_height()+1.5),
                    ha='center', va='bottom',
                    color= 'black')
plt.savefig('Results_Monolingual_mBERT.png', bbox_inches='tight')
plt.show()
```

```python
monolingual.to_csv('Monolingual_results_mBERT.csv', index=False)
```

4) Analysing results per distance

```python
average_results = {}

#filter out monolingual results and results with stdv > 10
lan_results_f1_fil_temp = lan_results_f1[lan_results_f1['LAN_1'] != lan_results_f1['LAN_2']].reset_index()

# filter out of results with high standard deviation over the 4 iterations -> not representive
lan_results_f1_fil = lan_results_f1_fil_temp[lan_results_f1_fil_temp['standard dev'] < 07.0].reset_index()

lan_results_f1_fil.loc[:,'common_features'] = pow(lan_results_f1_fil['DIST'],2)
#print(lan_results_f1_fil['common_features'].unique())
lan_results_f1_fil['common_features'] = lan_results_f1_fil['common_features'].apply(lambda x: 12 - int(x))
#print(lan_results_f1_fil['common_features'].unique())

for dist in lan_results_f1_fil['common_features'].unique():
    df_temp = lan_results_f1_fil[(lan_results_f1_fil['common_features'] == dist)]
    mean = np.mean(df_temp['mean'].values)
    #! the standard deviation of the language pair means
    std = np.std(df_temp['mean'].values)
    #! the standard deviation of 4 iterations represented as mean over all language pairs
    std_4 = np.mean(df_temp['standard dev'].values)
    
    
    average_results[dist] = [mean, std, std_4]

results_per_dist = pd.DataFrame.from_dict(average_results, orient = 'index', columns=['f1_mean_over_lan_pairs','f1_std_over_lan_pairs','f1_std_over_4_it_lan_pairs']).reset_index().sort_values(by = 'index', ascending=False)
results_per_dist = results_per_dist.rename(columns={'index' : 'Shared ES-Features', 'f1_mean_over_lan_pairs' : 'F1-Score Mean', 'f1_std_over_lan_pairs': 'F1-Score Standard Deviation','f1_std_over_4_it_lan_pairs':'F1-Score Standard Deviation over 4 Iter.'})
results_per_dist = results_per_dist.apply(lambda x: round(x,2))
#results_per_dist = results_per_dist['Standard Deviation over 4 Iterations in %'].apply(lambda x: round(x,2))
```

##### 5) Filtering for results with Tatar(not part of the pretrained languages)

```python
results_Tatar = lan_results_f1_fil[(lan_results_f1_fil['LAN_1']=='Tat') | (lan_results_f1_fil['LAN_2']=='Tat')].sort_values(by='DIST')
results_Tatar = results_Tatar.rename(columns={'common_features' : 'Shared ES-Features'})
results_Tatar
```

```python
Tatar = pd.merge(results_Tatar[['DIST','LAN_1','LAN_2','mean','standard dev','Shared ES-Features']], results_per_dist, on='Shared ES-Features')
```

```python
Tatar = Tatar.rename(columns={'F1-Score Mean':'dist. f1 mean', 'Shared ES-Features' : 'shared ES-features', 'F1-Score Standard Deviation':'dist. averg.f1 std dev','standard dev':'std (over 4 iter)','F1-Score Standard Deviation over 4 Iter.':'dist f1 std (over 4 iter)'})
Tatar
```

```python
Tatar = Tatar[['shared ES-features','DIST','LAN_1','LAN_2','mean','std (over 4 iter)','dist f1 std (over 4 iter)','dist. f1 mean','dist. averg.f1 std dev']]
```

```python
Tatar['\u0394 mean'] = Tatar['mean'] - Tatar['dist. f1 mean']
Tatar[['DIST','mean','std (over 4 iter)','\u0394 mean']] = Tatar[['DIST','mean','std (over 4 iter)','\u0394 mean']].apply(lambda x : round(x,2))
```
