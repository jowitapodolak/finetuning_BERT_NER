# Features: Comparison mBERT and XLMRoBERTa

```python
import pandas as pd

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mBERT = pd.read_csv('mBERT/Features_compressed_mBERT.csv', sep=',')
mBERT.head()
```

```python
XLMR = pd.read_csv('XLMR/Features_compressed_XLMR.csv', sep=',')
XLMR.head()
```

```python
temp1 = mBERT[['FEATURE','EXPERIM','FINT1_f1_mean','FINT2_f1_mean','rel. Diff']]
temp2 = XLMR[['FEATURE','EXPERIM','FINT1_f1_mean','FINT2_f1_mean','rel. Diff']]
comparison = pd.merge(temp1, temp2, on=['FEATURE','EXPERIM'])
```

```python
comparison = comparison.rename(columns={"FINT1_f1_mean_x": "FINT1_f1_mean_mBERT", "FINT1_f1_mean_y": "FINT1_f1_mean_XLMR","FINT2_f1_mean_x": "FINT2_f1_mean_mBERT", "FINT2_f1_mean_y": "FINT2_f1_mean_XLMR", "rel. Diff_x": "rel. Diff_mBERT", "rel. Diff_y": "rel. Diff_XLMR"})
comparison
```

```python
# Comparison f1-score fine-tuning 1 between mBERT and XLMR
temp = comparison[['FEATURE','EXPERIM']].values.tolist()
objects = []
for each in range(0,len(temp)):
    objects += ['feature' + str(int(temp[each][0])) + '_' + str(int(temp[each][1]))]

x = np.arange(len(objects))
y_mBERT = comparison['FINT1_f1_mean_mBERT'].values.tolist()
y_XLMR = comparison['FINT1_f1_mean_XLMR'].values.tolist()

fig, ax = plt.subplots(figsize=(15,25))
width = 0.3
opacity = 0.8

rects1 = ax.barh(x - width/2, y_mBERT, width, label='mBERT')
rects2 = ax.barh(x + width/2, y_XLMR, width, label='XLMR')

ax.set_xlabel('F1-Score (Mean over 4 Iterations) in %')
ax.set_ylabel('Feature and Experiment(1/2)')
ax.set_title('Fine-Tuning 1 - Comparison F1 Scores between mBERT and XLMR')
ax.set_yticks(x)
ax.set_xlim(xmax = 90, xmin = 60)
ax.set_yticklabels(objects, rotation=45)
ax.xaxis.grid(True, linestyle='--', which='major',
                   color='grey', alpha=.25)
ax.legend()

def autolabel(rects, count):
# attach some text labels
    for ii,rect in enumerate(rects):
        width = int(rect.get_width())

        height = rect.get_height()
        #print(height,width)
        yloc1=rect.get_y() + height /2.0
        yloc2=rect.get_y() + height /2.0
        # Shift the text to the right side of the right edge
        xloc1 = width + 1
        yloc2=yloc2+0.3
        # Black against white background
        clr = 'black'
        align = 'left'
        yloc1=rect.get_y() + height /2.0
        #print(xloc1,yloc1,yloc2)
        ax.text(xloc1,yloc1, '%s'% (round(count[ii],1)) + '%',horizontalalignment=align,
                         verticalalignment='center',color=clr,
                         clip_on=True)


autolabel(rects1, y_mBERT)
autolabel(rects2, y_XLMR)

fig.tight_layout()
plt.show()
fig.savefig('Comparison_mBERT_XLMR_f1scores_FT1.png')

# Comparison f1-score fine-tuning 2 between mBERT and XLMR
temp = comparison[['FEATURE','EXPERIM']].values.tolist()
objects = []
for each in range(0,len(temp)):
    objects += ['feature' + str(int(temp[each][0])) + '_' + str(int(temp[each][1]))]

x = np.arange(len(objects))
y_mBERT = comparison['FINT2_f1_mean_mBERT'].values.tolist()
y_XLMR = comparison['FINT2_f1_mean_XLMR'].values.tolist()

fig, ax = plt.subplots(figsize=(15,25))
width = 0.3
opacity = 0.8

rects1 = ax.barh(x - width/2, y_mBERT, width, label='mBERT')
rects2 = ax.barh(x + width/2, y_XLMR, width, label='XLMR')

ax.set_xlabel('F1-Score (Mean over 4 Iterations) in %')
ax.set_ylabel('Feature and Experiment(1/2)')
ax.set_title('Fine-Tuning 2 - Comparison F1 Scores between mBERT and XLMR')
ax.set_xlim(xmax = 90, xmin = 60)
ax.set_yticks(x)
ax.set_yticklabels(objects, rotation=45)
ax.xaxis.grid(True, linestyle='--', which='major',
                   color='grey', alpha=.25)
ax.legend()

def autolabel(rects, count):
# attach some text labels
    for ii,rect in enumerate(rects):
        width = int(rect.get_width())

        height = rect.get_height()
        #print(height,width)
        yloc1=rect.get_y() + height /2.0
        yloc2=rect.get_y() + height /2.0
        # Shift the text to the right side of the right edge
        xloc1 = width + 1
        yloc2=yloc2+0.3
        # Black against white background
        clr = 'black'
        align = 'left'
        yloc1=rect.get_y() + height /2.0
        #print(xloc1,yloc1,yloc2)
        ax.text(xloc1,yloc1, '%s'% (round(count[ii],1)) + '%',horizontalalignment=align,
                         verticalalignment='center',color=clr,
                         clip_on=True)


autolabel(rects1, y_mBERT)
autolabel(rects2, y_XLMR)

fig.tight_layout()
plt.show()
fig.savefig('Comparison_mBERT_XLMR_f1scores_FT2.png')
```

```python
# Comparison f1-score fine-tuning 2 between mBERT and XLMR
# IN PROCESS: Try to solve the issue with the width

objects = comparison['FEATURE'].unique().tolist()

Exp1 = comparison[comparison['EXPERIM'] == 1.0]
Exp2 = comparison[comparison['EXPERIM'] == 2.0]

x = np.arange(len(objects))
y1_mBERT = Exp1['FINT1_f1_mean_mBERT'].values.tolist()
y2_mBERT = Exp1['FINT2_f1_mean_mBERT'].values.tolist()
y3_mBERT = Exp2['FINT1_f1_mean_mBERT'].values.tolist()
y4_mBERT = Exp2['FINT2_f1_mean_mBERT'].values.tolist()
y1_XLMR = Exp1['FINT1_f1_mean_XLMR'].values.tolist()
y2_XLMR = Exp1['FINT2_f1_mean_XLMR'].values.tolist()
y3_XLMR = Exp2['FINT1_f1_mean_XLMR'].values.tolist()
y4_XLMR = Exp2['FINT2_f1_mean_XLMR'].values.tolist()

fig, ax = plt.subplots(figsize=(25,5))
width = 0.11
opacity = 0.8

rects1 = ax.bar(x - (3 * width + width/2), y1_mBERT, width, color = 'darkblue',edgecolor='black', hatch="/")
rects2 = ax.bar(x - (2 * width + width/2), y2_mBERT, width, label='mBERT_Experiment 1', color = 'darkblue',edgecolor='black')
rects3 = ax.bar(x - (width + width/2), y3_mBERT, width, color = 'royalblue',edgecolor='black', hatch="/")
rects4 = ax.bar(x - width/2, y4_mBERT, width, label='mBERT_Experiment 2', color = 'royalblue',edgecolor='black')

rects5 = ax.bar(x + width/2, y1_XLMR, width, color = 'chocolate',edgecolor='black', hatch="/")
rects6 = ax.bar(x + (width + width/2), y2_XLMR, width, label='XLMR_Experiment 1', color= 'chocolate',edgecolor='black')
rects7 = ax.bar(x + (2 * width + width/2), y3_XLMR, width, color = 'goldenrod',edgecolor='black', hatch="/")
rects8 = ax.bar(x + (3 * width + width/2), y4_XLMR, width, label='XLMR_Experiment 2', color= 'goldenrod',edgecolor='black')

ax.set_ylabel('F1-Score in %', fontsize=14)
ax.set_xlabel('Feature (Experiment 1 and 2 for mBERT/XLMR)', fontsize=16)
#ax.set_title('Fine-Tuning 2 - Comparison F1 Scores between mBERT and XLMR')
ax.set_xticks(x)
ax.set_ylim(ymin=65, ymax = 90)
ax.set_xticklabels(objects, fontsize=16)
ax.xaxis.grid(True, linestyle='--', which='major',
                   color='grey', alpha=.25)
ax.legend()

plt.grid(True, linestyle='--')
fig.tight_layout()
plt.show()
fig.savefig('Comparison_mBERT_XLMR_f1scores.png')
```
